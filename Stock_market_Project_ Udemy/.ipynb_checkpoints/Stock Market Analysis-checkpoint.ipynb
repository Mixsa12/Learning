```python
Tech_list = [ "AAPL", "GOOG", "AMD", "AMZN"]
```

```python



closing_df = pd.DataFrame()

for Tech in Tech_list:
    closing_df[Tech]=DataReader(Tech, "iex", start,end)['close']
```

```python
closing_df
```

```python
tech_rets = closing_df.pct_change() 
```

```python
tech_rets.head()
```

```python
sns.jointplot("GOOG", "GOOG", tech_rets, kind = 'scatter')
```

```python
sns.jointplot("GOOG", "AMD", tech_rets, kind = 'scatter')
```

```python
sns.pairplot(tech_rets.dropna())
```

```python
r_fig = sns.pairplot(tech_rets.dropna())
r_fig.map_upper(plt.scatter, color="purple" )
r_fig.map_lower(sns.kdeplot, cmap ="cool_d" )

r_fig.map_diag(plt.hist, bins= 50 )
```

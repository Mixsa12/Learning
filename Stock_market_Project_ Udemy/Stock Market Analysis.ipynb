```python
days= 365

dt = 1/days
mu= rets.mean()['GOOG']
sigma = rets.std()["GOOG"]
```

```python
def stock_monte_carlo(start_price, day, mu, sigma):
    price = np.zeros(days)
    price[0]= start_price 
    
    shock = np.zeros(days)
    drift = np.zeros(days)
    
    for x in range(1, days ):
        shock[x]= np.random.normal(loc=mu*dt, scale= sigma*np.sqrt(dt))
        
        drift[x]= mu *dt
        
        price[x]= price[x-1] +(price[x-1]*(drift[x]+shock[x]))
    return price
    
```

```python
# The Usual Suspects
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
import math

sns.set_style("whitegrid")
#ML libs
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics

import statsmodels.api as sm  
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x= iris.data
y = iris.target
```

```python
print(iris.DESCR)
```

```python
iris_data = DataFrame(x, columns=["sepal length","sepal width",
    "petal length","petal width"] )
```

```python
iris_target = DataFrame(y, columns= ["Species"])
```

```python
def flower(x):
    if x==0:
        return "Setosa"
    elif x ==1: 
        return "Versicolour"
    else:
        return "Virginica"
```

```python
iris_target["Species"]=iris_target["Species"].apply(flower)  
```

```python
iris_target.head()
```

```python
 iris = pd.concat([iris_data, iris_target], axis= 1)
```

```python
iris.head()
```

```python
sns.pairplot(iris, hue= "Species")
```

```python
sns.countplot(x="petal length", data = iris, hue= "Species")
```

```python
sns.countplot(x="sepal length", data = iris, hue= "Species")
```

```python
log_model= LogisticRegression()
```

```python
X_train, X_test, Y_train, Y_test= sklearn.cross_validation.train_test_split(x, y, test_size = 0.4, random_state=3)
```

```python
log_model.fit(X_train, Y_train)
```

```python
from sklearn import metrics 
```

```python
Y_pred = log_model.predict(X_test)
```

```python
print( metrics.accuracy_score(Y_test, Y_pred))
```

# K - Nearist Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn =  KNeighborsClassifier(n_neighbors= 6)
```

```python
knn.fit(X_train, Y_train)
```

```python
knY_pred = knn. predict(X_test)
```

```python
print( metrics.accuracy_score(Y_test, knY_pred))
```

```python
for x in range(1,21):
       #KNeighborsClassifier(n_neighbors= x).fit(X_train, Y_train)
       print(metrics.accuracy_score(Y_test, KNeighborsClassifier(n_neighbors= x).fit(X_train, Y_train).predict(X_test)))
```

```python
Acc = [ (metrics.accuracy_score(Y_test, KNeighborsClassifier(n_neighbors= x).fit(X_train, Y_train).predict(X_test))) for x in range(1,21)]
```

```python
Acc
```

```python
plt.plot( range(1,21), Acc)
plt.xlabel("K Value for KNN")
plt.ylabel("Test Accuracy")
plt.title("Aaccracy of K vaules for KNN Model on Iris Dataset")
```

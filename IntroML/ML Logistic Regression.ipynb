```python
# standard 
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
import math

sns.set_style("whitegrid")
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics

import statsmodels.api as sm  
%matplotlib inline
```

```python
df = sm.datasets.fair.load_pandas().data
```

```python
df.head()
```

```python
def affair_check(x):
    if x !=0:
        return 1 
    else:
        return 0
```

```python
df["Had_affair"] =df['affairs'].apply(affair_check)
```

```python
occ_dum = pd.get_dummies(df['occupation'])
```

```python
hus_occ_dum = pd.get_dummies(df['occupation_husb'])
```

```python
occ_dum
```

```python
occ_dum.columns= ['occ1', 'occ2','occ3', 'occ4','occ5', 'occ6']
```

```python
hus_occ_dum.columns= ['hocc1', 'hocc2','hocc3', 'hocc4','hocc5', 'hocc6']
```

```python
x= df.drop(['occupation','occupation','Had_affair'], axis = 1 )
```

```python
dummies= pd.concat([occ_dum,hus_occ_dum], axis= 1)
```

```python
x = pd.concat([x,dummies], axis = 1 )
```

```python
x.head()
```

```python
x= x.drop(["occ1","hocc1"], axis=1)
```

```python
x= x.drop('affairs', axis =1 )
```

```python
x.head()
```

```python
Y= df.Had_affair
```

```python
Y = np.ravel(Y)
Y
```

```python
log_model= LogisticRegression()
```

```python
log_model.fit(x,Y)
log_model.score(x,Y)
```
